```python
x=df.drop(['Churn', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaymentMethod'],axis=1).values
y=df['Churn'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

#Model 1

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dropout, BatchNormalization
```

```python
earlystop=EarlyStopping(monitor='val_loss', mode='min', patience=5)
```

```python
model = Sequential()
#input_layer and first hidden layer
model.add(Dense(10, kernel_initializer='uniform', input_shape=(10,)))
model.add(BatchNormalization())

model.add(Dense(100, activation='relu', kernel_initializer='uniform', kernel_regularizer='L1'))
model.add(BatchNormalization())

model.add(Dense(100,activation='relu', kernel_initializer='uniform', kernel_regularizer='L1'))
model.add(BatchNormalization())

model.add(Dense(100,activation='relu', kernel_initializer='uniform', kernel_regularizer='L1'))
model.add(BatchNormalization())

model.add(Dense(100,activation='relu', kernel_initializer='uniform', kernel_regularizer='L1'))
model.add(BatchNormalization())

model.add(Dense(100,activation='relu', kernel_initializer='uniform', kernel_regularizer='L1'))
model.add(BatchNormalization())

model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=x_train,y=y_train, validation_data=(x_test,y_test),batch_size=32, epochs=100)
```

```python
model.summary()
```

```python
losses_df = pd.DataFrame(model.history.history)
losses_df
```

```python
losses_df.plot()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
predictions=model.predict(x_test)
```

```python
predictions
```

```python
y_classes = ((predictions > 0.5)+0).ravel()
```

```python
y_classes
```

```python
confusion_matrix(y_classes,y_test)
```

```python
print(classification_report(y_test, y_classes))
```
